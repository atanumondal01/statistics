```python
import seaborn as sns
import numpy as np
import scipy.stats as st
import pylab
```

```python
df=sns.load_dataset("tips")
df
```

```python
import matplotlib.pyplot as plt
```

```python
def plot(df,features):
    plt.figure(figsize=(16,9))
    plt.subplot(1,2,1)
    sns.histplot(df[features],kde=True)
    plt.subplot(1,2,2)
    st.probplot(df[features],dist="norm",plot=pylab)
    plt.show()
    
```

```python
plot(df,"total_bill")
```
